```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

plt.style.use('fivethirtyeight')
```

```python
train = pd.read_csv('application_train.csv')
```

```python
train.shape
```

```python
bureau = pd.read_csv('bureau.csv')
bureau.head()
```

```python
bureau.shape
```

```python
test = pd.read_csv('application_test.csv')
```

```python
train['is_test'] = 0
test['is_test'] = 1
```

```python
train = pd.concat([train, test])
```

```python
train.shape
```

```python
previous_loan_counts = bureau.groupby('SK_ID_CURR', as_index=False).size()
previous_loan_counts
```

```python
previous_loan_counts = previous_loan_counts.rename(columns = {'size': 'previous_loan_counts'})
```

```python
train = train.merge(previous_loan_counts, on='SK_ID_CURR', how='left')
```

```python
train.shape
```

```python
train['previous_loan_counts'] = train['previous_loan_counts'].fillna(0)
```

```python
train.head()
```

```python
bureau.info()
```

```python
cols = ['CREDIT_ACTIVE', 'CREDIT_CURRENCY', 'CREDIT_TYPE']
```

```python
pd.get_dummies(bureau[cols[0]])
```

```python
for x in cols:
    dummies = pd.get_dummies(bureau[cols[0]], prefix=x)
    bureau = pd.concat([bureau, dummies], axis=1)
```

```python
bureau.drop(cols, axis=1, inplace=True)
```

```python
bureau_agg = bureau.drop(columns = ['SK_ID_BUREAU']).groupby('SK_ID_CURR', as_index = False).agg(['count', 'mean', 'max', 'min', 'sum'])
```

```python
bureau_agg
```

```python
agg = bureau_agg[['CREDIT_ACTIVE_Active', 'CREDIT_ACTIVE_Closed', 'CREDIT_ACTIVE_Sold']]
agg
```

```python
bureau_agg.columns
```

```python
bureau_agg.columns = [x[0] + '_' + x[1] for x in bureau_agg.columns]
```

```python
bureau_agg.columns
```

```python
bureau_agg.columns = np.hstack([['SK_ID_CURR'], bureau_agg.columns[1:]])
```

```python
bureau_agg.head()
```

```python
train = train.merge(bureau_agg, on='SK_ID_CURR', how='left')
```

```python
numerical = [x for x in train.columns if train[x].dtype != 'object']
```

```python
train[numerical].head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
lr = LogisticRegression()
```

```python
X = train[train['is_test'] == 0][numerical].fillna(0)
y = train[train['is_test'] == 0]['TARGET']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
y_test.shape
```

```python
lr.predict_proba(X_test)[:,1].shape
```

```python
roc_auc_score(y_test, lr.predict_proba(X_test)[:,1])
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model = knn.fit(X_train, y_train)
```

```python
knn_predictions = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, knn_predictions)
print(f'Accuracy: {accuracy}')
```
